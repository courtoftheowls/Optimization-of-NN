```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def interval_width(w):
    return ((1 / w) * np.log((w / 2) * (1 + np.sqrt(1 - 4 / w)) - 1)) - ((1 / w) * np.log((w / 2) * (1 - np.sqrt(1 - 4 / w)) - 1))

w_values = np.linspace(4, 1000, 1000-3)
#w_values = w_values[1:-1]
I_values = interval_width(w_values)

plt.figure(figsize=(10, 6))
plt.plot(w_values, I_values, label="Interval Width I = h^+(w) - h^-(w)")
plt.xlabel("Weight (w)")
plt.ylabel("Interval Width (I)")
plt.title("Interval Width I vs. Weight (w)")
plt.grid(True)
plt.legend()
plt.show()
```

The plot illustrates that as  w  increases, the interval width  I  decreases. In other words, as w→+∞ we observe I→0, telling us that the region susceptible to gradient explosion becomes negligibly small. This suggests that in deep networks with large weight values, gradient explosion is confined to a very limited range of input currents h. Therefore, this intensifies the vanishing gradient issue, further complicating the training of deep nuera networks.
